```python
# import the libraries

import numpy as np
import matplotlib.pyplot as plt
import glob
import cv2

from keras.models import Model, Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from keras.layers import BatchNormalization
from keras.applications.vgg16 import VGG16
import os
import seaborn as sns
```

```python
# Dataset path Name

dataset_hold = "E:/My Projects For PortFolio/AI CV Projects/medical_image_segmentation/model_training/Dataset"
dataset_normal = "E:/My Projects For PortFolio/AI CV Projects/medical_image_segmentation/model_training/Dataset/Normal"
dataset_cyst = "E:/My Projects For PortFolio/AI CV Projects/medical_image_segmentation/model_training/Dataset/Cyst"
```

```python
# display directory

print(os.listdir(dataset_hold))
```

```python
# size of the image

size = 256
```

```python
# list for holding images and labels

train_images = []
train_labels = []
```

```python
# labelling the images

path_dir = dataset_hold + "/*"
print(path_dir)
for directory_path in glob.glob(path_dir):
    label = directory_path.split("\\")[-1]
    print(label)
    
    for img_path in glob.glob(os.path.join(directory_path, "*.jpg")):
        print(img_path)
        print("/n")
        
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        img = cv2.resize(img, [size, size])
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        
        # append the images
        train_images.append(img)
        train_labels.append(label)
```

```python
# making the images as arrays

train_images = np.array(train_images)
train_labels = np.array(train_labels)
```

```python
# for test images

test_images = []
test_labels = []
```

```python
# labelling the images

path_test = "E:/My Projects For PortFolio/AI CV Projects/medical_image_segmentation/model_training/Dataset/test/*"
for directory_path in glob.glob(path_test):
    
    # label creation
    label = directory_path.split("\\")[-1]
    print(label)
    
    # reading the image
    for img_path in glob.glob(os.path.join(path_test, "*jpg")):
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        img = cv2.resize(img, [size, size])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        # appending to the list
        test_images.append(img)
        test_labels.append(label)
```

```python
# array of the test images

test_images = np.array(test_images)
test_labels = np.array(test_labels)
```

```python
# pre-processing

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# train labels
le.fit(train_labels)
train_labels_encoded = le.transform(train_labels)

# test labels
le.fit(test_labels)
test_labels_encoded = le.transform(test_labels)
```

```python
# segregating the train and test data

x_train, y_train, x_tst, y_test = train_images, train_labels_encoded, test_images, test_labels_encoded
```

```python
# making the image data equivalent for model

x_train, x_test = train_images / 255.0, test_images / 255.0
```

```python
# VGG Model

VGG_Model = VGG16(
            weights = "imagenet",
            include_top = False,
            input_shape = [size, size, 3]
        )
```

```python
# layers

for layer in VGG_Model.layers:
    layer.trainable = False
```

```python
print(VGG_Model.summary())
```

```python
# feature Extraction

feature_extractor = VGG_Model.predict(x_train)
```

```python
features = feature_extractor.reshape(feature_extractor.shape[0], -1)
X_for_training = features
```

```python
import xgboost as xgb
```

```python
model = xgb.XGBClassifier()
```

```python
model.fit(X_for_training, y_train)
```

```python
x_test_feature = VGG_Model.predict(x_test)
x_test_features = x_test_feature.reshape(x_test_feature.shape[0], -1)
```

```python
prediction = model.predict(x_test_features)
prediction = le.inverse_transfrom(prediction)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
acc_mat = accuracy_matrix(test_labels, prediction)
print(acc_mat)
```
